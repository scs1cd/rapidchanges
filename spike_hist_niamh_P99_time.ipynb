```python
def cdfLSMOD(adtdt,advgpdt,advdmdt,adfdt,atime,var,n):
   
    cdfLSMOD = np.zeros(n)

    for i in range(0,n,1):
        t = atime[i,:]
        if var==1:
            Q = adtdt[i,:]
            label = "$\partial \\hat{\\mathbf{B}}/ \partial t$ ($^{\\circ}$ yr$^{-1}$)"
        elif var==2:
            Q = advgpdt[i,:]
            label = "$\partial \\hat{\\mathbf{P}}_V/ \partial t$ ($^{\\circ}$ yr$^{-1}$)"
        elif var==3:
            Q = abs(advdmdt[i,:])
            label = "$\partial P_V/ \partial t$ ($ZAm^2$ yr$^{-1}$)"
        elif var==4: 
            Q = abs(adfdt[i,:]/1000.0) # microTesla
            label = "$\partial B/ \partial t$ (${\\mu}$T yr$^{-1}$)"
      
        array = Q[Q>0]
        pts = np.sort(array)
        
        muX       = np.mean(np.log(array))
        sigmaX    = np.std(np.log(array)) 
        logNdist1 = lognorm(sigmaX, scale=np.exp(muX), loc=0)
        
        cdf       = logNdist1.cdf(pts)
        cdfindex  = np.argmax(cdf >= 0.99)
        #print(cdfindex)
        P99index = cdfindex
        P99value = pts[P99index]
        cdfLSMOD[i] = P99value
        
       
    return cdfLSMOD
```

```python
def GGF_LSMODcdf(adtdt,advgpdt,advdmdt,adfdt,atime,var,n,cdf):
   
    alltop1 = []
    alltime1 = []

    for i in range(0,n,1):
        t = atime[i,:]
        if var==1:
            Q = adtdt[i,:]
            label = "$\partial \\hat{\\mathbf{B}}/ \partial t$ ($^{\\circ}$ yr$^{-1}$)"
        elif var==2:
            Q = advgpdt[i,:]
            label = "$\partial \\hat{\\mathbf{P}}_V/ \partial t$ ($^{\\circ}$ yr$^{-1}$)"
        elif var==3:
            Q = abs(advdmdt[i,:])
            label = "$\partial P_V/ \partial t$ ($ZAm^2$ yr$^{-1}$)"
        elif var==4: 
            Q = abs(adfdt[i,:]/1000.0) # microTesla
            label = "$\partial B/ \partial t$ (${\\mu}$T yr$^{-1}$)"
      
        array = Q[Q>0]
        pts = np.sort(array)
        if np.max(Q) >= cdf[i]:
            indices = np.where(Q>cdf[i])
            keep=np.zeros(len(indices[0]))
            times=np.zeros(len(indices[0]))
            for j in range(0,len(indices[0]),1):
                keep[j] = Q[indices[0][j]]
                times[j] = t[indices[0][j]]
        else: 
            indices = np.nan
            keep = np.nan
            times = np.nan
        #print(indices)
        #print(cdf[i])
        #print(np.max(Q))
        
        alltop1.append(keep)
        alltime1.append(times)
       
    return alltop1, alltime1
```
